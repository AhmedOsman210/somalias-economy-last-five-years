```python
import matplotlib.pyplot as plt

# Somalia's real GDP growth rates (2019-2023)
years = [2019, 2020, 2021, 2022, 2023]
gdp_growth = [3.6, -2.6, 3.3, 2.4, 4.2]

plt.figure(figsize=(8, 5))
plt.plot(years, gdp_growth, marker='o', linestyle='-', color='b', label='Real GDP Growth (%)')

# Adding title and labels
plt.title("Somalia's Real GDP Growth (2019-2023)")
plt.xlabel("Year")
plt.ylabel("GDP Growth Rate (%)")
plt.grid(True)
plt.legend()

# Annotating data points
for i, value in enumerate(gdp_growth):
    plt.annotate(f"{value}%", (years[i], gdp_growth[i]), textcoords="offset points", xytext=(0, 8), ha='center')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {'Age': [15, 18, 20, 22, 24, 30, 35],
        'Height': [150, 160, 165, 170, 175, 180, 185]}

df = pd.DataFrame(data)

# 1. Histogram - Distribution of Age
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=5, edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 2. Bar Chart - Category comparison (Height)
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 15, 30, 10, 50]
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Category Comparison')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# 3. Pie Chart - Distribution of Ages
labels = ['15-20', '21-25', '26-30', '31-35']
sizes = [2, 2, 1, 2]  # Assume this distribution for simplicity
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['red', 'blue', 'green', 'yellow'])
plt.title('Age Distribution Pie Chart')
plt.show()

# 4. Statistical Features - Mean, Median, Mode, Standard Deviation
mean = df.mean()
median = df.median()
mode = df.mode().iloc[0]
std_dev = df.std()

print("Statistical Features:")
print("Mean:\n", mean)
print("Median:\n", median)
print("Mode:\n", mode)
print("Standard Deviation:\n", std_dev)

# 5. Pair Plot - Visualizing relationships between variables
sns.pairplot(df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample economic growth data (GDP growth rates over 5 years)
years = [2019, 2020, 2021, 2022, 2023]
gdp_growth = [2.5, -0.5, 4.0, 3.2, 5.1]  # Example growth rates in percentage

# Create a DataFrame
economic_data = pd.DataFrame({
    'Year': years,
    'GDP Growth (%)': gdp_growth
})

# 1. Line Chart - Economic Growth Trend
plt.figure(figsize=(10, 6))
plt.plot(economic_data['Year'], economic_data['GDP Growth (%)'], marker='o', color='b', linestyle='-', linewidth=2)
plt.title('Economic Growth Trend (GDP Growth Rate)')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.grid(True)
plt.show()

# 2. Bar Chart - Comparing Economic Growth Across Years
plt.figure(figsize=(10, 6))
plt.bar(economic_data['Year'], economic_data['GDP Growth (%)'], color='green')
plt.title('GDP Growth Comparison')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.show()

# 3. Pie Chart - Share of Economic Growth in Different Sectors (Example)
sectors = ['Agriculture', 'Industry', 'Services']
growth_share = [15, 40, 45]  # Hypothetical share of GDP growth by sector

plt.figure(figsize=(8, 8))
plt.pie(growth_share, labels=sectors, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('GDP Growth Share by Sector')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: amount of money each state received over the last 5 years (in millions)
# (This is just sample data, replace it with actual data)
data = {
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Puntland': [50, 55, 60, 65, 70],
    'Somaliland': [60, 65, 70, 75, 80],
    'Jubaland': [40, 45, 50, 55, 60],
    'South West': [30, 35, 40, 45, 50],
    'Galmudug': [20, 25, 30, 35, 40]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the total amount each state received over the last 5 years
total_received = df.drop(columns=['Year']).sum()

# Calculate the average amount received each year per state
average_received = df.drop(columns=['Year']).mean()

# Rank states based on total received (descending order)
ranked_total_received = total_received.sort_values(ascending=False)

# Rank states based on average received (descending order)
ranked_average_received = average_received.sort_values(ascending=False)

# Print the total and average received with ranks
print("Total Money Received by Each State (Ranked):")
print(ranked_total_received)
print("\nAverage Money Received per Year by Each State (Ranked):")
print(ranked_average_received)

# Plotting the total money received by each state
plt.figure(figsize=(10, 6))
ranked_total_received.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title('Total Money Received by Each State (Ranked over 5 Years)')
plt.xlabel('State')
plt.ylabel('Total Money Received (Millions)')
plt.xticks(rotation=45)
plt.show()

# Plotting the average money received per year by each state
plt.figure(figsize=(10, 6))
ranked_average_received.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title('Average Money Received per Year by Each State (Ranked)')
plt.xlabel('State')
plt.ylabel('Average Money Received (Millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Volume of Kenyan khat imported each year (in tons) and average price per ton (in millions of Somali Shillings)
data = {
    'Year': [2020, 2021, 2022, 2023],
    'Volume Imported (Tons)': [50, 55, 60, 65],  # Example volume (Tons)
    'Price per Ton (Millions)': [2.5, 2.6, 2.7, 2.8]  # Average price per ton (in Millions)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the total cost for each year
df['Total Cost (Millions)'] = df['Volume Imported (Tons)'] * df['Price per Ton (Millions)']

# Calculate total spending over the 4 years
total_spending = df['Total Cost (Millions)'].sum()

# Print the total cost and breakdown by year
print("Spending on Kenyan Khat over the Last 4 Years:")
print(df)
print(f"\nTotal Spending by the Somali Government on Kenyan Khat over the Last 4 Years: {total_spending} Million Somali Shillings")

# Plotting the total spending per year
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Total Cost (Millions)'], color='teal')
plt.title('Somali Government Spending on Kenyan Khat (Last 4 Years)')
plt.xlabel('Year')
plt.ylabel('Total Spending (Millions of Somali Shillings)')
plt.show()
```
